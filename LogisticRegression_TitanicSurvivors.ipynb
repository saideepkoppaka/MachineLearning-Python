```python
#Creating a function to deal with missing age fields
```

```python
titanic.head(2)
```

```python
def emptyage(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
titanic['Age'] = titanic[['Age','Pclass']].apply(emptyage, axis=1)
```

```python
sns.heatmap(titanic.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
titanic.drop('Cabin',axis=1,inplace = True)
```

```python
titanic.dropna(inplace=True)
```

```python
titanic.head()
```

```python
Gender = pd.get_dummies(titanic['Sex'],drop_first=True)
```

```python
Gender
```

```python
Embark = pd.get_dummies(titanic['Embarked'],drop_first=True)
```

```python
Par = pd.get_dummies(titanic['Parch'],drop_first=True)
```

```python
titanic = pd.concat([titanic,Gender,Embark,Par],axis =1 )
```

```python
titanic.head(2)
```

```python
titanic.drop(['Sex', 'Embarked', 'Name','Ticket'], axis=1, inplace=True)
```

```python
titanic.drop('PassengerId',axis=1,inplace=True)
```

```python
titanic.head()
```

```python
X = titanic.drop('Survived',axis=1)
```

```python
Y = titanic['Survived']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Logmodel = LogisticRegression()
```

```python
Logmodel.fit(X_train,Y_train)
```

```python
predictions = Logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print classification_report(Y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print confusion_matrix(Y_test, predictions)
```
